```python
from numpy import *
from matplotlib.pyplot import *
% matplotlib inline
```

```python
#Datos:
t11 = array([0.0,30.0,60.0,90.0,120.0,150.0,180.0,210.0,240.0,270.0,300.0,306.0])
v11 = array([5.02,3.97,3.15,2.50,2.00,1.60,1.28,1.02,0.82,0.65,0.53,0.5])

t12 = array([0.0,30.0,60.0,90.0,120.0,150.0,180.0,210.0,240.0,270.0,300.0,307.0])
v12 = array([5.00,3.96,3.16,2.52,2.01,1.61,1.28,1.02,0.83,0.66,0.53,0.5])

t13 = array([0.0,30.0,60.0,90.0,120.0,150.0,180.0,210.0,240.0,270.0,300.0,309.0])
v13 = array([5.05,3.99,3.17,2.53,2.02,1.61,1.29,1.03,0.82,0.66,0.53,0.5])
```

```python
#Grafico Escala "Normal":
scatter(t11,v11,color='red')
scatter(t12,v12,color='blue')
scatter(t13,v13,color='green')
title('Voltaje v/s Tiempo')
xlabel(r'$t$')
ylabel(r'$V$')
grid()
```

```python
#Grafico Escala Semi-Logaritmica:
scatter(t11,v11,color='red')
scatter(t12,v12,color='blue')
scatter(t13,v13,color='green')
title('ln(V) v/s Tiempo')
xlabel(r'$t$')
ylabel(r'$ln(V)$')
grid()
yscale('log')
```

```python
#Ajuste de Curvas:
z1 = polyfit(t11,log(v11),1)
z2 = polyfit(t12,log(v12),1)
z3 = polyfit(t13,log(v13),1)
```

```python
T11 = linspace(0,310,1000)
V11 = e**(z1[1] + z1[0]*T11)

T12 = linspace(0,310,1000)
V12 = e**(z2[1] + z2[0]*T12)

T13 = linspace(0,310,1000)
V13 = e**(z3[1] + z3[0]*T13)
```

```python
#Grafico Escala Semi-Logaritmica:
scatter(t11,v11,color='red')
scatter(t12,v12,color='blue')
scatter(t13,v13,color='green')
plot(T11,V11,color='red')
plot(T12,V12,color='blue')
plot(T13,V13,color='green')
title('ln(V) v/s Tiempo')
xlabel(r'$t$')
ylabel(r'$ln(V)$')
grid()
yscale('log')
```

```python
#Grafico Escala "Normal":
scatter(t11,v11,color='red')
scatter(t12,v12,color='blue')
scatter(t13,v13,color='green')
plot(T11,V11,color='red')
plot(T12,V12,color='blue')
plot(T13,V13,color='green')
title('Voltaje v/s Tiempo')
xlabel(r'$t$')
ylabel(r'$V$')
grid()
```

```python
#Coeficientes:
print(z1,z2,z3)
```

Obtenemos asi las siguientes funciones:
    \begin{equation}
        V_1 = e^{-0.00749t +  1.59971} = 4.95159e^{-0.00749t}
    \end{equation}
    \begin{equation}
        V_2 = e^{-0.00747t +  1.59935} = 4.94981e^{-0.00747t}
    \end{equation}
    \begin{equation}
        V_3 = e^{-0.00748t +   1.60568} = 4.98124e^{-0.00748t} 
    \end{equation}

Recordemos que la solucion para este circuito es:
    \begin{equation}
        V = V_0 e^{-\frac{1}{RC}t} = V_0 e^{-\frac{1}{\tau_C}t}
    \end{equation}

Por lo que podemos asociar la "amplitud" de cada ajuste, a $V_{0}$ para ese respectivo ajuste. Notemos que este no necesariamente concuerda con el valor real para $V_{0}$ (valor inicial)

Ademas, podemos asociar los coeficientes de los exponentes de cada ajuste a $-\frac{1}{\tau_{C}}$, por lo que:
    \begin{equation}
        -\frac{1}{\tau_1} = -0.00749 \Rightarrow \tau_1 = 133.51134
    \end{equation}
    \begin{equation}
        -\frac{1}{\tau_2} = -0.00722 \Rightarrow \tau_2 = 133.86880
    \end{equation}
    \begin{equation}
        -\frac{1}{\tau_3} = -0.00723 \Rightarrow \tau_3 = 133.68983
    \end{equation}
    
Comparando con los valores teoricos:
    \begin{equation}
        \tau_1 = \tau_2 = \tau_3 = RC = 117.6409
    \end{equation}
